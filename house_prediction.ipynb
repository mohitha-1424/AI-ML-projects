```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = {
    'Area': [600, 800, 1000, 1200, 1500, 1800, 2000, 2200, 2500, 3000],
    'Price': [3000000, 4000000, 5000000, 6000000, 7500000,
              9000000, 10000000, 11000000, 12500000, 15000000]
}
df = pd.DataFrame(data)
X = df[['Area']]
y = df['Price']
model = LinearRegression()
model.fit(X, y)
new_area = [[1600]]
predicted_price = model.predict(new_area)
print("Predicted Price for", new_area[0][0], "sq.ft house = ₹", round(predicted_price[0], 2))
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.scatter(new_area, predicted_price, color='green', s=100, label='Predicted House')
plt.title("House Price Prediction using Linear Regression")
plt.xlabel("House Area (sq.ft)")
plt.ylabel("House Price (₹)")
plt.legend()
plt.grid(True)
plt.show()
```
