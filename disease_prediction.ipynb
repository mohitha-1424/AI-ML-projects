```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
data = {
    'Age': [25, 35, 45, 50, 60, 40, 55, 65, 30, 48],
    'BloodPressure': [120, 130, 140, 150, 160, 135, 145, 170, 125, 155],
    'Cholesterol': [180, 200, 220, 240, 260, 210, 230, 270, 190, 250],
    'HeartRate': [72, 75, 80, 85, 90, 78, 82, 95, 74, 88],
    'Disease': [0, 0, 1, 1, 1, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
X = df[['Age', 'BloodPressure', 'Cholesterol', 'HeartRate']]
y = df['Disease']
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
model.fit(X, y)
age = int(input("Enter Age: "))
bp = int(input("Enter Blood Pressure: "))
chol = int(input("Enter Cholesterol: "))
hr = int(input("Enter Heart Rate: "))
prediction = model.predict([[age, bp, chol, hr]])
print("\nPrediction Result")
if prediction[0] == 1:
    print("Disease Detected")
else:
    print("No Disease Detected")
```
