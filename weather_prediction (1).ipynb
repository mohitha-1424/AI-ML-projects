```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
data = {
    'Temperature': [30, 25, 20, 18, 35, 28, 22, 24, 32, 19],
    'Humidity':    [70, 80, 90, 95, 60, 75, 85, 78, 65, 92],
    'WindSpeed':   [10, 15, 20, 25, 8, 12, 18, 16, 9, 22],
    'Weather': ['Sunny', 'Rainy', 'Rainy', 'Rainy',
                'Sunny', 'Cloudy', 'Cloudy', 'Cloudy',
                'Sunny', 'Rainy']
}
df = pd.DataFrame(data)
X = df[['Temperature', 'Humidity', 'WindSpeed']]
y = df['Weather']
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y_encoded)
temp = float(input("Enter Temperature (°C): "))
humidity = float(input("Enter Humidity (%): "))
wind = float(input("Enter Wind Speed (km/h): "))
prediction = model.predict([[temp, humidity, wind]])
weather = encoder.inverse_transform(prediction)
print("\nPredicted Weather:", weather[0])
```
